```python
# Load all CSV files from the database/data directory into a dict of DataFrames
from pathlib import Path
import pandas as pd

data_dir = Path('/home/dataopske/Desktop/AcademIQ/database/data')

# find all .csv files
csv_files = sorted(data_dir.glob('*.csv'))

# read into a dict: key = stem (filename without suffix), value = DataFrame
dfs = {f.stem: pd.read_csv(f) for f in csv_files}

# also create top-level variables for convenience (e.g., students -> dfs['students'])
for name, df in dfs.items():
    globals()[name] = df

# print a summary of loaded DataFrames
print(f"Found {len(dfs)} CSV files in {data_dir}")
for name, df in dfs.items():
    print(f"- {name}: shape={df.shape}")
```
