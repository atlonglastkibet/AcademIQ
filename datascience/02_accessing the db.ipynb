```python
tables = [
    "buses", "class_attendance", "classes", "exams",
    "fee_payments", "route_stops", "routes", "schools",
    "student_parents", "student_results", "student_transport",
    "student_transport_logs", "students", "subjects",
    "terms", "users"
]

data = {}

for table in tables:
    response = supabase.table(table).select("*").execute()
    data[table] = response.data  # store each table's data in a dictionary

print(f"âœ… Loaded {len(data)} tables successfully!")
```

```python
# View all columns for each table
print("\nðŸ“‹ Table Schemas (Column Names):")
for table_name, table_data in data.items():
    if table_data:  # Skip if empty
        columns = list(table_data[0].keys())  # Get columns from first row
        print(f"\n{table_name}: {', '.join(columns)}")
    else:
        print(f"\n{table_name}: [Empty table - no columns to display]")
```
